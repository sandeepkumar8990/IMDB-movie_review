```python
#importing the libraries.
import os
import re
import nltk
import random
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
import os
from sklearn.externals import joblib
import pickle
```

```python
# Reading the data from train data which contains positive and negative reviews
imdb_folder='D:/imdb reviews/aclImdb'
def get_reviews(data_folder='/train'):
    reviews = []
    
    for sentiment in ['/neg/','/pos/']:
        path=imdb_folder+data_folder+sentiment
        for filename in sorted(os.listdir(path)):
            with open(path + filename,encoding='utf-8') as f:
                review=f.read()
                reviews.append(review)
                
                
                
    return reviews   
```

```python
#creating an object of train reviews.
train_reviews=get_reviews()
```

```python
#checkign the length of reviews.
len(train_reviews)
```

```python
#checkign the first review in the data.
train_reviews[0]
```

```python
type(train_reviews)
```

```python
#filling white spaces instead of \n
train_reviews=[re.sub("\n","",i)for i in train_reviews]
```

```python
#creating Target variable with 1's and 0's. 1 for negative and 0 for positive.
target= [1 if i < 12500 else 0 for i in range(25000)]
```

```python
#converting into a data frame.
import pandas as pd
train = pd.DataFrame([train_reviews,target]).T
train.columns = ["review","Target"]
```

```python
train.head()
```

```python
train.shape
```

```python
train.Target.value_counts()
```

```python
train.Target.unique()
```

```python
#converting data type of target variable to intiger.
train.Target=train.Target.astype(str).astype('int64')
```

```python
#tokenizing the sentence into sinle tokens.
train['review'] = train['review'].str.lower().str.split()
```

```python
#removing the stop words from the data.
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
```

```python
print(train['review'][0])
```

```python
train['review']=train['review'].apply(lambda x: str([item for item in x if item not in stopwords]))
```

```python
print(train['review'][0])
```

```python
#remoing the patterns like \/,[\|]
pattern = r'\.|\"|\,|\[|\]|\''
train['review']=train['review'].apply(lambda x: re.sub(pattern,'',str(x)))
```

```python
#visualising most popular words from all reviews
all_words = ' '.join([text for text in train['review']])
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

wordcloud = WordCloud(width=800, background_color='white',height=500, random_state=21, max_font_size=110).generate(all_words)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
#visualising the popular words in negative reviews.
normal_words =' '.join([text for text in train['review'][train['Target'] == 1]])

wordcloud = WordCloud(width=800,background_color='white', height=500, random_state=21, max_font_size=110).generate(normal_words)
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
# spliting the data into train and validation.
from sklearn.model_selection import train_test_split 

X_train, x_valid, y_train, y_valid = train_test_split(train['review'], train['Target'], train_size=0.7, random_state=1234)
```

```python
print(X_train.shape)
print(x_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
X_train.head()
```

```python
#converting the train data into sturctured form by using TF-IDF Vectorizer.
from sklearn.feature_extraction.text import TfidfVectorizer

#Geting doc-term matrix on train data
tf=TfidfVectorizer(binary=True)
                   

tdm_train = tf.fit_transform(X_train)
```

```python
#Geting  doc-term matrix for validation data
tdm_valid = tf.transform(x_valid)
```

```python
print(tdm_train.shape)
print(tdm_valid.shape)
```

```python
type(tdm_train)
```

```python
#model 1 
from sklearn.naive_bayes import MultinomialNB
navie=MultinomialNB()
navie.fit(tdm_train,y_train)
```

```python
from sklearn.metrics import classification_report
train_pred1=navie.predict(tdm_train)
valid_pred1=navie.predict(tdm_valid)

##checking the classification report and accuracy of the model
print("Train_Classification Report: \n\n", classification_report(y_train,train_pred1))
print("\n\nvalid_Classification Report: \n\n", classification_report(y_valid,valid_pred1))
```

```python
#model 2
from sklearn.linear_model import LogisticRegression

log=LogisticRegression()
log.fit(tdm_train,y_train)
```

```python
train_pred2=log.predict(tdm_train)
valid_pred2=log.predict(tdm_valid)

##checking the classification report and accuracy of the model
print("Train_Classification Report: \n\n", classification_report(y_train,train_pred2))
print("\n\nvalid_Classification Report: \n\n", classification_report(y_valid,valid_pred2))
```
